```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd

# Example with only numeric data
num_features = X_train.select_dtypes(include=['number']).columns

pipelines = {
    'KNN': Pipeline([
        ('scaler', StandardScaler()),
        ('model', KNeighborsRegressor())
    ]),
    'RF': Pipeline([
        ('scaler', StandardScaler()),
        ('model', RandomForestRegressor(random_state=42))
    ]),
    'Boosting': Pipeline([
        ('scaler', StandardScaler()),
        ('model', GradientBoostingRegressor(random_state=42))
    ])
}

# Fit all models
for name, pipe in pipelines.items():
    pipe.fit(X_train[num_features], y_train)

# Calculate MSE
mse = pd.DataFrame(columns=['train', 'test'], index=pipelines.keys())
for name, pipe in pipelines.items():
    y_train_pred = pipe.predict(X_train[num_features])
    y_test_pred = pipe.predict(X_test[num_features])

    mse.loc[name, 'train'] = mean_squared_error(y_train, y_train_pred) / 1e3
    mse.loc[name, 'test'] = mean_squared_error(y_test, y_test_pred) / 1e3

print(mse)
```

```python
fig, ax = plt.subplots()
mse.sort_values(by='test', ascending=False).plot(kind='barh', ax=ax, zorder=3)
ax.grid(zorder=0)
```

```python
import numpy as np
import pandas as pd

def single_row_predictions(model_dict, X_test, y_test, row=0):
    """
    model_dict: {'KNN': knn, 'RF': RF, 'Boosting': boosting}
    X_test: the SAME preprocessed test features used for model training
    y_test: matching target (Series or array)
    row: which row index to predict (int)
    """
    # --- get the row in a consistent 2D shape ---
    if isinstance(X_test, pd.DataFrame):
        row_df = X_test.iloc[[row]].copy()          # keep DataFrame, 2D
        row_df = row_df.fillna(0)                   # no NaNs
        base_df = row_df
        base_arr = None
    else:
        arr = np.asarray(X_test)
        base_arr = arr[row:row+1]                   # 2D slice
        base_arr = np.nan_to_num(base_arr)          # no NaNs
        base_df = None

    # --- y_true value ---
    if isinstance(y_test, pd.Series):
        y_true_val = float(y_test.iloc[row])
    else:
        y_true_val = float(np.asarray(y_test)[row])

    out = {'y_true': [y_true_val]}

    # --- predict with each model robustly ---
    for name, model in model_dict.items():
        # If we have a DataFrame and the model knows its training feature names, align columns
        if base_df is not None and hasattr(model, 'feature_names_in_'):
            need_cols = list(model.feature_names_in_)
            # add any missing columns as 0 (rare but safe)
            for col in need_cols:
                if col not in base_df.columns:
                    base_df[col] = 0
            Xi = base_df.loc[:, need_cols].to_numpy(dtype=np.float64)
        elif base_df is not None:
            # model doesn't expose feature_names_in_; just pass numeric array
            Xi = base_df.to_numpy(dtype=np.float64)
        else:
            # X_test was an array to begin with
            Xi = np.asarray(base_arr, dtype=np.float64)

        # predict
        yhat = model.predict(Xi)
        yhat_val = float(np.asarray(yhat).ravel()[0])
        out['prediksi_' + name] = [round(yhat_val, 1)]

    return pd.DataFrame(out)


```
